```python
overall_team_stats = pd.DataFrame(
{'Total Matches Played': ipl3["team1"].value_counts() + ipl3["team2"].value_counts(), 
'Toss Won': ipl3["toss_winner"].value_counts(), 'Total Won': ipl3["winner"].value_counts(), 
 'Total Lost': ((ipl3["team1"].value_counts() + ipl3["team2"].value_counts()) - ipl3["winner"].value_counts())})
overall_team_stats.sort_values(by="Total Won", ascending=False)
```

```python
ax = overall_team_stats["Toss Won"].sort_values().plot(kind="barh", title="Number of Toss win in IPL", figsize=(12,8))# graph for the total toss win
```

```python
overall_team_stats['Win Ratio'] = overall_team_stats['Total Won'] * 100 / overall_team_stats['Total Matches Played']
overall_team_stats['Loss Ratio'] = overall_team_stats['Total Lost'] * 100 / overall_team_stats['Total Matches Played']
overall_team_stats.round(2).sort_values(by="Win Ratio", ascending = False)
```

```python
wins = ipl3["toss_winner"] == ipl3["winner"]
ax = wins.value_counts().plot(kind="pie", autopct='%.0f%%') #Pie chart of the comparision for team which has won toss and won match also
```

```python
ipl3["winner"][ipl3["win_by_wickets"]!=0].mode() # most time the team win by wickets we can say by chasing most succesful chasing team
```

```python
wins = ipl3["toss_decision"][ipl3["toss_winner"]==ipl3["winner"]]
wins.value_counts().plot(kind="pie", autopct='%1.1f%%')# comparision chart on the basis of decision of toss 
```

```python
# toss and match wins by toss_winner
toss_winner_as_winner = ipl3[ipl3['winner'] == ipl3['toss_winner']].groupby(['toss_winner'])['winner'].count()
# total toss wins by toss_winner
total_toss_winner = ipl3.groupby(['toss_winner'])['winner'].count()
win_per_on_toss_win = toss_winner_as_winner / total_toss_winner * 100
win_per_on_toss_win = win_per_on_toss_win.to_frame()
win_per_on_toss_win["lost_per_on_toss_win"] = 100 - win_per_on_toss_win['winner']
win_per_on_toss_win.rename(columns={'winner': 'win_per_on_toss_win'}, inplace=True)
win_per_on_toss_win.round(2).sort_values(by="win_per_on_toss_win", ascending=False)
```

```python
# partenship between batsman and non striker
ipl[ipl["batting_team"] == "MI"].groupby(["batsman", "non_striker"])["total_runs"].sum().sort_values(ascending = False).to_frame()
```

```python
super_overs = ipl[ipl["is_super_over"] == 1].groupby("batsman")["total_runs"].sum().sort_values(ascending = False).head()
super_overs.plot(kind="bar", figsize=(15, 5)) # Most successful batsman in the super over
```

```python
super_overs = ipl[(ipl["is_super_over"] == 1) & (ipl["player_dismissed"].notnull())].groupby("bowler")["player_dismissed"].count().sort_values(ascending = False).head()
super_overs.plot(kind="bar", figsize=(15, 5))# Most successful bowler in the super over
```

```python
orange_cap_players = ipl4.groupby(["season"]).apply(lambda x: (
 x.groupby(["batsman"]).sum()
 .sort_values('batsman_runs', ascending=False)).head(1))
orange_cap_players["batsman_runs"].to_frame() # orange cap player by season
```

```python
purple_cap_players = ipl4[ipl4["player_dismissed"].notnull()].groupby(["season"]).apply(lambda x: (
 x.groupby(["bowler"]).count()
 .sort_values("player_dismissed", ascending=False)).head(1))
purple_cap_players["player_dismissed"].to_frame()# Purple cap winner
```
